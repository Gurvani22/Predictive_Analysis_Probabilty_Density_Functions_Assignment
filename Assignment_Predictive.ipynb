```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize

# Loading Dataset
df = pd.read_csv("data.csv", encoding="latin1")
# Transforming Data
x = df["no2"].dropna().astype(float).values
r = 102317080

ar = 0.05 * (r % 7)
br = 0.3 * ((r % 5) + 1)


z = x + ar * np.sin(br * x)


def negative_log_likelihood(params, data):
    lam, mu, c = params


    if lam <= 0 or c <= 0:
        return np.inf

    p = c * np.exp(-lam * (data - mu)**2)


    p = np.clip(p, 1e-12, None)

    return -np.sum(np.log(p))


lam0 = 1 / (2 * np.var(z))
mu0 = np.mean(z)
c0 = 1.0

initial_params = [lam0, mu0, c0]


result = minimize(
    negative_log_likelihood,
    initial_params,
    args=(z,),
    method="L-BFGS-B"
)

lam_hat, mu_hat, c_hat = result.x

# Output Parameters
print("Lambda =", lam_hat)
print("Mu     =", mu_hat)
print("C      =", c_hat)
```
